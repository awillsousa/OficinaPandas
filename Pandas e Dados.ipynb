### Apagando dados repetidos

```python
# Carregar os dataframes
df_plan1 = pd.read_excel('dados/planilha1.xlsx')
df_plan2 = pd.read_excel('dados/planilha2.xlsx')
df_plan3 = pd.read_excel('dados/planilha3.xlsx')

# Concatenando 3 dataframes
df_plan1_2_3 = pd.concat([df_plan1, df_plan2, df_plan3])

df_plan1_2_3
```

```python
# Para apagar dados duplicados em DataFrame
df_plan1_2_3.drop_duplicates()
```

```python
# Para apagar dados duplicados em DataFrame
# e deixar somente dados não repetidos
df_plan1_2_3.drop_duplicates(keep=False)
```

```python
# Apagando os valores repetidos baseando em uma coluna específica
df_plan1_2_3.drop_duplicates(subset=["Município", "Modalidade"], keep=False)
```

```python
# Se queremos descobrir os Órgão Concedentes e as Modalidades com menor Valor Global (Vl_Global)
df_plan1_2_3_ordenado = df_plan1_2_3.sort_values(by=["Órgão_Concedente", "Modalidade", "Vl_Global"])

df_plan1_2_3_ordenado

#df_plan1_2_3_ordenado[["Órgão_Concedente", "Modalidade", "Vl_Global"]]
```

```python
df_menores_valores = df_plan1_2_3_ordenado.drop_duplicates(subset=["Órgão_Concedente", "Modalidade"], keep='first')

df_menores_valores[["Órgão_Concedente", "Modalidade", "Vl_Global"]]
```

```python
# E se quisermos os maiores?
df_plan1_2_3_ordenado = df_plan1_2_3.sort_values(by=["Órgão_Concedente", "Modalidade", "Vl_Global"])

df_maiores_valores = df_plan1_2_3_ordenado.drop_duplicates(subset=["Órgão_Concedente", "Modalidade"], keep='last')

df_maiores_valores[["Órgão_Concedente", "Modalidade", "Vl_Global"]]
```

```python
# Também pode apagar linhas que contenham células vazias ou com dados faltando
df_plan1_2_3.dropna()
```
